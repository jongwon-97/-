```python
import pandas as pd

# CSV 파일 읽기
file1 = pd.read_csv('2018딸기생육전처리.csv', encoding='utf-8')
file2 = pd.read_csv('2018딸기재배환경전처리.csv', encoding='utf-8')

# 첫 번째 파일의 조사일자를 날짜 형식으로 변환
file1['조사일자'] = pd.to_datetime(file1['조사일자']).dt.date

# 두 번째 파일의 측정시간을 datetime 형식으로 변환하고 날짜 추출
file2['측정시간'] = pd.to_datetime(file2['측정시간'])
file2['측정일자'] = file2['측정시간'].dt.date

# 숫자형 데이터만 선택하되, '농가명'과 '작기'는 제외
numeric_columns = file2.select_dtypes(include=['float64', 'int64']).columns
numeric_columns = numeric_columns.drop(['농가명', '작기'])

# 두 번째 파일에서 '농가명', '작기'를 그대로 유지하고 그룹화
file2_grouped = file2.groupby(['측정일자', '농가명', '작기'])[numeric_columns].mean().round(2).reset_index()

# 첫 번째 파일에 있는 조사일자들로 두 번째 파일을 필터링
filtered_file2 = file2_grouped[file2_grouped['측정일자'].isin(file1['조사일자'])]

# 결과 확인
filtered_file2
```

```python
import pandas as pd

# 첫 번째 파일과 두 번째 파일 병합 (조사일자와 측정일자를 기준)
# merged_data = pd.merge(file1, filtered_file2, left_on='조사일자', right_on='측정일자', how='inner')
merged_data = pd.read_csv("딸기데이터병합.csv",encoding="utf-8")
# 필요한 열만 선택 (화방별착과수와 환경 관련 열)
columns_for_corr = ['화방별착과수','온도_내부', '상대습도_내부','초장','엽수','엽장','엽폭','엽병장','관부직경','화방번호']

# 상관관계 계산
correlation_matrix = merged_data[columns_for_corr].corr()

# 결과 확인
correlation_matrix
```

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np
merged_data = merged_data.replace([np.inf, -np.inf], np.nan)
merged_data = merged_data.dropna()  # 결측값이 있는 행 제거
# 독립 변수 (환경 변수들)

X = merged_data[['온도_내부', '상대습도_내부','초장','엽수','엽장','엽폭','엽병장','관부직경','화방번호']]

# 종속 변수 (화방별착과수)
y = merged_data['화방별착과수']

# 상수항 추가
X = sm.add_constant(X)

# 다중 회귀 모델 피팅
model = sm.OLS(y, X).fit()

# 결과 출력
model.summary()
```
