```python
import pandas as pd

# CSV 파일을 데이터프레임으로 불러오기
data = pd.read_csv("2020_환경_통합.csv",encoding="euc-kr")
data= data.drop(columns= ["온도_외부","풍속_외부","풍향_외부","누적일사량_외부","강우감지",'일사량_외부','잔존CO2','토양온도'])
data = data[data["품목"]== "딸기"]
# 데이터 확인 (첫 몇 줄 출력)
data
```

```python
data.to_csv("2020딸기내부환경.csv",index=False)
```

```python
import pandas as pd

# CSV 파일을 데이터프레임으로 불러오기
data = pd.read_csv("딸기_2020_생육_통합.csv",encoding="euc-kr")
data['화방별착과수'] = data['화방별착과수'].fillna(0)
data= data.drop(columns= ["화방번호","액아구분"])
data
```

```python
columns_to_fill = ['초장', '엽수', '엽장', '엽폭', '엽병장', '관부직경']

# 농가명, 조사일자, 개체번호로 그룹화하여 결측값을 채움
for col in columns_to_fill:
    data[col] = data.groupby(['농가명', '조사일자', '개체번호'])[col].transform(lambda x: x.fillna(x.mean()))

# 결과 확인
data
```

```python
data.to_csv("2020딸기생육.csv",index=False)
```
