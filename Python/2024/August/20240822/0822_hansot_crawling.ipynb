```python
from selenium import webdriver #셀레니움 모듈을 활용한 웹크롤링 기능 가져옴. 
from webdriver_manager.chrome import ChromeDriverManager #크롬브라우저 설치를 해주는 드라이버. 
from selenium.webdriver.chrome.service import Service #최신버전의 크롬드라이버 설치 지원.

from selenium.webdriver.common.by import By # html 관련 요소들을 선택할 수 있는 모듈.

from selenium.webdriver.chrome.options import Options 

from bs4 import BeautifulSoup as bs #html코드 해석(번역).
import pandas as pd #행렬구조의 2차원 테이블 형태로 데이터를 저장,분석.시각화를 지원하는 모듈.
import time #딜레이 시간 설정.

import os
from urllib.request import urlretrieve

#ChromeDriverManager를 이용해서 크롬드라이버를 설치 .
serv = Service(executable_path=ChromeDriverManager().install())
#sevice를 이용해서 최신버전의 크롬브라우저를 만든 후 driver 변수에 할당.
driver = webdriver.Chrome(service=serv)

driver.get("https://www.hsd.co.kr/menu/menu_list")
time.sleep(3)

try:
    for i in range(10):
        btn = driver.find_element(By.CSS_SELECTOR,"a.c_05").click()
        time.sleep(3)
except:
    print("더이상 표시할 메뉴가 없습니다.")

title=driver.find_elements(By.CSS_SELECTOR,"div.item-text>h4")
price=driver.find_elements(By.CSS_SELECTOR ,".item-price")    
# title=driver.find_elements(By.CSS_SELECTOR,"div.item-text>h4")
# price=driver.find_elements(By.CSS_SELECTOR ,".item-price")

#이미지를 저장할 폴더(디렉토리 생성)
if not os.path.isdir("C:/hansot"):
    os.mkdir("C:/hansot")

title_list=[]
price_list=[]
src_list=[]
#menuList_560 > div > div.item-img > img
img=driver.find_elements(By.CSS_SELECTOR,"div.item-img img")

print("-"*50)
print("메뉴","                                  가격")
print("-"*50)
for i in range(len(title)):
    title_list.append(title[i].text.strip().replace("[8월 할인메뉴]",""))
    price_list.append(price[i].text.strip())
    src_list.append(img[i].get_attribute("src"))
    print(f"{title_list[i]:<30}{price_list[i]:>10}")
dic={"메뉴명":title_list,"가격":price_list}
han=pd.DataFrame(dic)
han.set_index("메뉴명",inplace=True)
han.to_csv("한솥메뉴_euc.kr.csv",encoding="euc-kr")
han.to_csv("한솥메뉴_utf-8.kr.csv",encoding="utf-8")
han.to_csv("한솥메뉴utf-8-sig.kr.csv",encoding="utf-8-sig")

for i in range(len(src_list)):
    urlretrieve(src_list[i],"C:/hansot/"+title_list[i]+".jpg" ) #urlretrieve(이미지 주소,저장할경로 및 파일명)

driver.quit()
```
