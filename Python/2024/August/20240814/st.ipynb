```python
# import sys
# sys.setrecursionlimit(5000)
# block_cipher = None
# a = Analysis(['run_mvp.py'],
#              pathex=['I:\\1ST_MEETING\\CODE\\DEEP_EYE_GPU\\MVP_2'],
#              binaries=[],
#              datas=[],
#              hiddenimports=[],
#              hookspath=[],
#              runtime_hooks=[],
#              excludes=[],
#              win_no_prefer_redirects=False,
#              win_private_assemblies=False,
#              cipher=block_cipher,
#              noarchive=False)
# pyz = PYZ(a.pure, a.zipped_data,
#              cipher=block_cipher)
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd
import pyautogui as py
def stock_data(page):
    # 네이버 금융 시세 페이지의 기본 URL
    url = (f'https://finance.naver.com/sise/sise_market_sum.naver?sosok=0&page={page}') 
                                                    #?sosok=0&page= 이부분에서 sosok=0은 코스피 시장 page는 페이지를 가져온다
    res = req.get(url)
    html = res.text 
    soup = bs(html, 'lxml')
    
    # 주식 정보를 저장할 빈 리스트 생성
    data = []
    
    # 주식 정보를 포함한 모든 행(tr)을 선택
    stocks = soup.select('table.type_2 tbody tr')
    
    # 각 행에서 데이터를 추출
    for stock in stocks:
        # 주식명 추출 (td 태그 내의 a 태그)
        name = stock.select_one('td a')
        # 현재가, 전일비, 등락률을 추출 (각각 td 태그의 nth-child 사용) 
        pr = stock.select_one('td:nth-child(3)')  # 3번째 칸의 현재가
        #contentarea > div.box_type_l > table.type_2 > tbody > tr:nth-child(2) > td:nth-child(3)
        ch = stock.select_one('td:nth-child(4)')  # 4번째 칸의 전일비
        ra = stock.select_one('td:nth-child(5)')  # 5번째 칸의 등락률

        # 필요한 데이터가 모두 존재할 때만 처리
        if name and pr and ch and ra:
            # .strip()은 문자열 양 끝의 공백 문자(스페이스, 줄 바꿈 등)를 제거함
            # HTML 태그에서 텍스트를 추출할 때 앞뒤에 공백이 포함될 수 있어서 이를 제거하기 위해 사용함
            data.append({
                '종목명': name.text.strip(),  # 종목명에서 앞뒤 공백 제거 후 추가 
                '현재가': pr.text.strip(),  # 현재가에서 앞뒤 공백 제거 후 추가
                '전일비': ch.text.strip(),  # 전일비에서 앞뒤 공백 제거 후 추가
                '등락률': ra.text.strip()  # 등락률에서 앞뒤 공백 제거 후 추가
            })
    
    # 리스트 데이터를 pandas DataFrame으로 변환 (표 형식으로 처리하기 위함)
    df = pd.DataFrame(data)
    return df
# 사용자가 입력한 페이지 번호를 받아옴.
page =py.prompt("보고싶은 페이지를 입력해주세요")
# page = int(input("페이지 번호를 입력하세요 (1, 2 등): "))

# 입력받은 페이지에 해당하는 주식 데이터를 가져옴.
df = stock_data(page)

# df파일을 csv 파일로 전환 3가지 방법.
df.set_index("종목명",inplace=True)   #첫번째 줄을 종목명으로 설정.
df.to_csv("주식차트_utf8.csv",encoding="utf8")
df.to_csv("주식차트_euc-kr.csv",encoding="euc-kr")
df.to_csv("주식차트_utf-8-sig.csv",encoding="utf-8-sig")
print("데이터가 파일로 저장되었습니다.")
print("="*100)
print(df)
```
