```python
#네이버 뉴스 타이틀 여러개 가져오기
import requests as req
from bs4 import BeautifulSoup as bs

res=req.get("https://news.naver.com/section/105")

html=res.text
soup=bs(html,"lxml")
#_SECTION_HEADLINE_LIST_ynl9b > li:nth-child(1) > div > div > div.sa_text > a > strong
titles=soup.select(".sa_text_strong") #class 를가져왔다 리스트가 여러개여서 계속 바뀌기 때문에 
counter = 1
for title in titles:
    print(f"{counter}.{title.text}")
    counter += 1
```

```python
#네이버 뉴스 타이틀 여러개 가져오기
import requests as req
from bs4 import BeautifulSoup as bs

res=req.get("https://search.naver.com/search.naver?ssc=tab.news.all&where=news&sm=tab_jum&query=%EC%82%BC%EC%84%B1%EC%A0%84%EC%9E%90")

html=res.text
soup=bs(html,"lxml")
#_SECTION_HEADLINE_LIST_ynl9b > li:nth-child(1) > div > div > div.sa_text > a > strong
titles=soup.select(".news_tit") #class 를가져왔다 리스트가 여러개여서 계속 바뀌기 때문에 
i= 1
for title in titles:
    print(f"{i}.{title.text}\n{title.attrs["href"]}") #attrs [ 속성  ]  href의 속성을 가져온다 
    i+= 1
```

```python
#삼성전자의 현재주가를 화면에 출력한다
import requests as req
from bs4 import BeautifulSoup as bs
import pyautogui

codes=['122870','005930','128940','003490']
print("========주식 현황판=========")
for cd in codes:
    res=req.get(f"https://finance.naver.com/item/sise.naver?code={cd}")
    html=res.text
    soup=bs(html,"lxml")
        
    price =soup.select_one("#_nowVal ").text #id가 _nowVal로 지정된 태그로 가서 문자부분만 선탣
    company =soup.select_one(".wrap_company a").text
    titles =soup.select_one("#content > div.section.inner_sub > div:nth-child(1) > table > tbody > tr:nth-child(1) > th:nth-child(1)").text
    print(company,"\n",titles,price)
```

```python
import pandas as pd  #판다스 모듈 
import requests as req
from bs4 import BeautifulSoup as bs

head ={"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/127.0.0.0 Safari/537.36"}
res=req.get("https://www.melon.com/chart/index.htm", headers=head)
html=res.text
soup=bs(html,"lxml")

song = soup.select("div.ellipsis.rank01 > span > a") #list의 형태로 결과가 저장됨.
singer = soup.select("div.ellipsis.rank02 > a") #list의 형태로 결과가 저장됨.
#lst50 > td:nth-child(6) > div > div > div.ellipsis.rank02 > a
#lst50 > td:nth-child(6) > div > div > div.ellipsis.rank02 > span > a
#conts > div.wrap_dtl_atist > div > div.wrap_atist_info > p
song_list = []   #빈 리스트 생성
singer_list=[]   #빈 리스트 생성
rank_list=[]

i=1
for i in range(len(song)):
    song_list.append(song[i].text)
    singer_list.append(singer[i].text)
    rank_list.append(i+1)
    print(rank_list[i],song_list[i],singer_list[i])
#pandas 모듈 사용
dic={"순위": rank_list,"노래 제목": song_list, "가수 이름": singer_list} #dic[키:값]으로 만듦 

melon=pd.DataFrame(dic) #데이터 프레임 melon에 값 저장
melon.set_index("순위",inplace=True)
melon.to_csv("멜론차트_utf8.csv",encoding="utf8")
melon.to_csv("멜론차트_euc-kr.csv",encoding="euc-kr")
melon.to_csv("멜론차트_utf-8-sig.csv",encoding="utf-8-sig")  #깨질때 한글로 인코딩 방법 3가지
```
